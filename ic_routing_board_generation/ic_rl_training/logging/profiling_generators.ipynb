### RUNNING WITH BATCHES

```python
uniform_generator = UniformRandomGenerator(10, 5)
random_walk_generator = JaxRandomWalkGenerator(10, 5)
seed_extension_generator = RandomSeedGenerator(10, 5)
random_walk_parallel_generator = ParallelRandomWalkGenerator(10, 5)
batch_size = 100
keys = jax.random.split(jax.random.PRNGKey(0), batch_size)
```

Uniform Generator

```python
#UNIFORM
uniform_generator_call = jax.jit(jax.vmap(uniform_generator.__call__))
%time jax.block_until_ready(uniform_generator_call(keys))  # compilation
```

```python
%timeit jax.block_until_ready(uniform_generator_call(keys))  # using cache
```

```python
random_walk_generator_call = jax.jit(jax.vmap(random_walk_generator.__call__))
%time jax.block_until_ready(random_walk_generator_call(keys))  # compilation
```

```python
%timeit jax.block_until_ready(random_walk_generator_call(keys))  # using cache
```
